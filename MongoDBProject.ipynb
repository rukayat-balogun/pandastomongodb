```python
database = client.customerOrder
collection = database.order_details
def create_documents():

    omrrts = []
    firstNames = []
    lastNames= []
    gender = []
    city = []
    country = []

    for i , j in df.iterrows():
        omrrts.append(j.omrrts)
        firstNames.append(j['First Name'])
        lastNames.append(j['Last Name'])
        gender.append(j['Gender'])
        city.append(j['City'])
        country.append(j['Country'])

    docs = []
        
    for omrrts, firstNames, lastNames, gender, city ,country in zip(omrrts, firstNames, lastNames, gender, city ,country):
        doc = {'omrrts':omrrts, 'First Name':firstNames, 'Last Name':lastNames, 'City':city, 'Country': country}
        #person_collection.insert_one(doc)
        docs.append(doc)
        
    collection.insert_many(docs)
```

```python
create_documents()
```
